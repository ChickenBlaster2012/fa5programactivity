```python
def get_total(one_price, how_many):
  the_sum = one_price * how_many
  return the_sum

print("")

item_name = input("What item did you bring?: ")
cost_string = input("How much does your item cost?: ")
count_string = input("How many did you bring?: ")

# Remove any non-numeric characters (like '$') before converting to float
cost_number = float(cost_string.replace('$', ''))
count_number = int(count_string)

my_final_cost = get_total(cost_number, count_number)

print(f"Item: {item_name}")
print(f"Cost per item: ${cost_number:.2f}")
print(f"Quantity: {count_number}")
print(f"Total cost: ${my_final_cost:.2f}")
```
